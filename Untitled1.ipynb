W = funcion lineal con el minimo error cuadratico medio

tenemos que

(X.W - Y)^2 = (Y-XW).T @ (Y-XW) = Y.T@Y - W.T@X.T@Y - Y.T@X@W + W.T@X.T@X@W

derivada(Y.T@Y - W.T@X.T@Y - Y.T@X@W + W.T@X.T@X@W) = -2@X.t@Y + 2@X.T@XW = 0


-2@X.T@Y + 2@X.T@XW = 0

W  = inv(X.T @ X) @ X.T @ Y





```python
import numpy as np

#libreria matematica

import matplotlib.pyplot as plt

#libreria para graficar datos

from sklearn.datasets import load_boston

boston = load_boston()

print(boston.DESCR)
```

```python
X = np.array(boston.data[:, 5])

Y = np.array(boston.target)


```

```python
trapito
plt.scatter(X, Y,alpha=0.3)


X = np.array([np.ones(506), X]).T

print(X)


W = np.linalg.inv(X.T @ X) @ X.T @ Y

plt.plot([4, 9], [W[0] + W[1] * 4, W[0] + W[1] * 9 ], c="red")

plt.show()
```
